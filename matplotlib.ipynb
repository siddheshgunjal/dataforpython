```python
style.use('ggplot')
x = [3,4,5,6]
y = [4,6,2,4]

x2 = [1,2,3,4]
y2 = [7,8,5,3]

plt.plot(x,y, color='green', label = '1st line', linewidth='5')
plt.plot(x2,y2, color='blue', label = '2nd line', linewidth='5')

plt.title('1st graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()

plt.show()
```

```python
x = [1,3,5,7]
y = [4,6,2,4]

x2 = [2,4,6,8]
y2 = [7,8,5,3]

plt.bar(x,y)
plt.bar(x2,y2)

plt.title('2nd graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
x = [1,3,5,7]
y = [4,6,2,4]

x2 = [2,4,6,8]
y2 = [7,8,5,3]

plt.bar(x,y, color="red", label = 'one')
plt.bar(x2,y2, color = 'green', label = 'two')

plt.legend()

plt.title('2nd graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
x = [1,3,5,7]
y = [4,6,2,4]

x2 = [2,4,6,8]
y2 = [7,8,5,3]

plt.scatter(x,y, color="red", label = 'one')
plt.scatter(x2,y2, color = 'green', label = 'two')

plt.legend()

plt.title('2nd graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
data = [4,6,3,2]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['c', 'm', 'r', 'b']

plt.pie(data,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow=True,
        explode=(0,0,0.1,0))
plt.title('pie chart')
plt.show()
        
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('data1.csv')
df=df.values
#time series vs reservoir levels(ft) graph
sns.set_style('darkgrid')
plt.plot(df[:,0],df[:,1],label="")
#plt.plot(df[:,0],df[:,2])
plt.xlabel('Time Series')
plt.ylabel('Reservoir Levels(ft)')
plt.title('Dialy Bhakhra Reservoir Levels for past 20 years')
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure()
ax1=plt.subplot(2,1,1)
line1,=ax1.plot(df[:,0],df[:,2],label='inflow')
line2,=ax1.plot(df[:,0],df[:,3],label='discharge',color='#ff7f0eff')
ax1.set_ylabel('Inflow & Discharge(Cusecs)',fontsize=8)



plt.subplot(2,1,2)
line3,=plt.plot(df[:,0],df[:,1],label='Reservoir Levels(ft)')
plt.xlabel('Time Series',fontsize=18)
plt.ylabel('Reservoir Levels(ft)',fontsize=8)

ax1.legend((line1,line2),('Inflow', 'Discharge'))
#ax2.legend()
plt.show()
```
